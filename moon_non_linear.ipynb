Hear i am making the random moon  with the help of moon_make function which gives me a 2d non-linear function and after doing that we are going to check for the kernal funtion and how they seperates the data.



```
%matplotlib inline
```

```
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)

plt.figure(figsize=(8,6))

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)

plt.title('A nonlinear 2Ddataset')
plt.ylabel('y coordinate')
plt.xlabel('x coordinate')

plt.show()
```

### 1.   Linear pca

### hear is the visulization of 1st two pca components.

```
#implimentation of classical pca on this 2d non linear Dataset we have

from sklearn.decomposition import PCA, KernelPCA

linear_pca = PCA(n_components=2)
X_lpca = linear_pca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_lpca[y==0, 0], X_lpca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_lpca[y==1, 0], X_lpca[y==1, 1], color='blue', alpha=0.5)

plt.title('First 2 principal components after Linear PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

visulizing the first pca1 component after doing PCA on this Dataset

```
import numpy as np
linear_pca = PCA(n_components=1)
X_lpca = linear_pca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_lpca[y==0, 0], np.zeros((50,1)), color='red', alpha=0.5)
plt.scatter(X_lpca[y==1, 0], np.zeros((50,1)), color='blue', alpha=0.5)

plt.title('First principal component after Linear PCA')
plt.xlabel('PC1')

plt.show()
```

As we can see above our linear pca function does not seperate this dataset.

### 2.   RBF kernal pca on our dataset.
###  hear gamma value depedns on the dataset and we can find the optimal gamma value by doing Grid search and another method but they are also from themselves are big topic so we are using gamma=15 which is used in many dataset to find a good results.



```
from sklearn.decomposition import KernelPCA

scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_skernpca[y==0, 0], X_skernpca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], X_skernpca[y==1, 1], color='blue', alpha=0.5)

plt.text(-0.48, 0.35,'gamma = 15', fontsize=12)
plt.title('First 2 principal components after RBF Kernel PCA via scikit-learn')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```
scikit_kpca = KernelPCA(n_components=1, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_skernpca[y==0, 0], np.zeros((50,1)), color='red', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], np.zeros((50,1)), color='blue', alpha=0.5)
plt.text(-0.48, 0.007, 'gamma = 15', fontsize=12)
plt.title('First principal component after RBF Kernel PCA')
plt.xlabel('PC1')
plt.show()
```
